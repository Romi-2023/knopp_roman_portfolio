## Na podstawie 150 próbek trzech gatunków Irysa przeprowadzimy analizę wielkości oraz szerokości łodygi jak i płatka.

```python
# Importujemy potrzebne biblioteki do analizy które pozwolą nam zwizualizować dane w przystępny sposób.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Odczytujemy dane.

```python
df = pd.read_csv('25__iris.csv', sep=",")
df
```

# Wyświetlamy podstawowe statystyki.

```python
df.describe ()
```

# Jako że są tyko trzy typy tej rośliny sprawdzamy czy dane są dla każdej klasy są tej samej wartości.

```python
df['klasa (class)'].value_counts()
```

## Jako że mamy równo po 50 sztuk danych dla każdej klasy. Przystępujemy do analizy wizualnej żeby informacje były jak najbardziej szczegółowe nie wybieramy losowych tylko analizujemy wszystko.

```python
grouped_df = df.groupby ('klasa (class)').mean()

# Budujemy wykres słupkowy dla długości, szerokości płatka i kielich wedłóg kalsy.
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Dane do porównania długości kielicha.
grouped_df['długość kielicha (sepal length)'].plot(kind='bar', ax=axes[0, 0], color=['red', 'green', 'blue'])
axes[0, 0].set_title('Średnia kielicha wedłóg klasy ')
axes[0, 0].set_ylabel('Długość kielicha (cm)')

# Dane do porównania szerokości kielicha.
grouped_df['szerokość kielicha (sepal width)'].plot(kind='bar', ax=axes[0, 1], color=['red', 'green', 'blue'])
axes[0, 1].set_title('Szerokość kielicha wedłóg klasy')
axes[0, 1].set_ylabel('Szerokość kielicha (cm)')

# Dane do porównania długości płatka.
grouped_df['długość płatka (petal length)'].plot(kind='bar', ax=axes[1, 0], color=['red', 'green', 'blue'])
axes[1, 0].set_title('Długość płatka według płatka')
axes[1, 0].set_ylabel('Długość płatka (cm)')

# Danae do porównania szerokości płatka.
grouped_df['szerokość płatka (petal width)'].plot(kind='bar', ax=axes[1, 1], color=['red', 'green', 'blue'])
axes[1, 1].set_title('Szerokość płatka według klasy')
axes[1, 1].set_ylabel('Szerokość płatka (cm)')

plt.tight_layout()
```

## Już po pierwszym wykresie widać że najmniejszy jest Iris-versicolor a największt Iris-virginica.

# Przedstawiamy kolejną wizualizację która uwidacznia te twierdzenie.

```python
plt.figure(figsize=(10, 6))
for species in df['klasa (class)'].unique():
    subset = df[df['klasa (class)'] == species]
    plt.scatter(subset['długość kielicha (sepal length)'], subset['szerokość kielicha (sepal width)'], label=species)

plt.title('Długość kielicha vs Szerokość kielicha')
plt.xlabel('Dłudość kielicha (cm)')
plt.ylabel('Szerokość kielicha (cm)')
plt.legend(title='Gatunek')
plt.grid(True)
```

# Czas na sprawdzenie wartości odstających w celu sprawdzenia spójności danych.

```python
sns.set(style="whitegrid")

# przygotowywujemy kolejną wizualizację
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Boxplot dla długości kielicha.
sns.boxplot(x='klasa (class)', y='długość kielicha (sepal length)', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Długość kielicha wedłóg klasy')

# Boxplot dla szerokości kielicha.
sns.boxplot(x='klasa (class)', y='szerokość kielicha (sepal width)', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Szerokość kielicha wedłóg klasy')

# Boxplot dla długości płatka.
sns.boxplot(x='klasa (class)', y='długość płatka (petal length)', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Długość płatka wedłóg klasy')

# Boxplot dla szerokości płatka.
sns.boxplot(x='klasa (class)', y='szerokość płatka (petal width)', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Szerokość płatka wedłóg klas')

# Adjust layout
plt.tight_layout()
```

## Widzimy że dane są spójne z małymi odchyłami. Mimo tych małych wartości odstających informacje są wiarygodne.

# Przedstawiamy kolejny wykres procentowy

```python
grouped_data = df.groupby('klasa (class)').mean()

# Tworzymy wykres kołowy dla długości kielicha.
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(grouped_data['długość kielicha (sepal length)'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Średnia długość kielicha wedłóg klasy')

# Tworzymy wykres kołowy dla szerokości kielicha .
plt.subplot(1, 2, 2)
plt.pie(grouped_data['szerokość kielicha (sepal width)'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Średnia szerokość kielicha wedłóg klasy')

# Tworzymy wykres kołowy dla długości płatka.
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.pie(grouped_data['długość płatka (petal length)'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Średnia długość płatka wedłóg klasy')

#Tworzymy wykres kołowy dla szerokości płatka.
plt.subplot(1, 2, 2)
plt.pie(grouped_data['szerokość płatka (petal width)'], labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.title('Średnia szerokość płatka wedłóg klasy')

plt.tight_layout()
```

## Ten wykres wpełni obrazuje nam wielkości poszczególnych Irysów.
